```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def fcc_lattice(n):
    lattice = []
    for i in range(3):
        for j in range(3):
            for k in range(3):
                if (i + j + k) % 2 == 0:
                    lattice.append([i, j, k])
    return np.array(lattice)

def bcc_lattice(n):
    lattice = []
    for i in range(n):
        for j in range(n):
            for k in range(n):
                if i%2==0 and j%2==0 and k%2==0:
                  lattice.append([i, j, k])
                if i%2!=0 and j%2!=0 and k%2!=0:
                  lattice.append([i, j, k])
    return np.array(lattice)

def plot_lattice(lattice, ax, color='b', r=0.5):
    for point in lattice:
        ax.scatter(point[0], point[1], point[2], color=color, s=100)
        u = np.linspace(0, 2 * np.pi, 100)
        v = np.linspace(0, np.pi, 100)
        x = r * np.outer(np.cos(u), np.sin(v)) + point[0]
        y = r * np.outer(np.sin(u), np.sin(v)) + point[1]
        z = r * np.outer(np.ones(np.size(u)), np.cos(v)) + point[2]
        ax.plot_surface(x, y, z, color=color, alpha=1)

fig = plt.figure(figsize=(10, 5))

# FCC lattice

ax1 = fig.add_subplot(121, projection='3d')
ax1.set_title('FCC Packing')
fcc = fcc_lattice(3)
plot_lattice(fcc, ax1)
"""
# BCC lattice
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_title('BCC Packing')
bcc = bcc_lattice(3)
plot_lattice(bcc, ax2, color='r')
"""
plt.tight_layout()
plt.show()
```
